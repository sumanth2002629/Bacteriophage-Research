```python
import pandas as pd
ds = pd.read_csv("/content/drive/MyDrive/Bacteriophage_Research/rbp_embeddings_prott5.csv")
```

```python
DNA_seq = ds["Nucleotide Sequence"]
len(DNA_seq)
```

```python
import torch
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
# tokenizer.to("cuda")
model = AutoModel.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
```

```python
DNA_embeddings = []
model = model.to("cuda")

for i in range(len(DNA_seq)):

  dna = DNA_seq[i]

  inputs = tokenizer(dna, return_tensors = 'pt')["input_ids"]

  inputs = inputs.to("cuda")

  with torch.no_grad():
    hidden_states = model(inputs)[0] # [1, sequence_length, 768]

  # embedding with mean pooling
  embedding_mean = torch.mean(hidden_states[0], dim=0)
  DNA_embeddings.append(embedding_mean) # expect to be 768

  inputs.detach().cpu()
  hidden_states.detach().cpu()



DNA_embeddings_tensor = torch.stack(DNA_embeddings)
torch.save(DNA_embeddings_tensor, "/content/embeddings_DB2.pt")
# print(DNA_embeddings)

# DNA_embeddings = torch.tensor(DNA_embeddings)

# print
# embedding with max pooling
# embedding_max = torch.max(hidden_states[0], dim=0)[0]
# print(embedding_max.shape) # expect to be 768
```
